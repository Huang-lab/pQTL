```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
cancer_type = 'BRCA'
# cancer_type = 'CRC'
```

```python
cancer = pd.read_csv('../../../Huang_lab_data/CPTAC2retrospective/' + 
                      cancer_type + '/' + cancer_type + '_clinical.txt', sep="\t")
```

```python
cancer.index = cancer['Unnamed: 0']
cancer = cancer.drop(['Unnamed: 0'], axis=1)
```

```python
cancer.head()
```

```python
stage_to_slice = cancer.loc['PATHOLOGIC_STAGE'][cancer.loc['PATHOLOGIC_STAGE']
                                              .astype(str).str.slice(start=-1).isin(['A', 'B', 'C'])]
stage_no_need_to_slice = cancer.loc['PATHOLOGIC_STAGE'][~cancer.loc['PATHOLOGIC_STAGE']
                                                      .astype(str).str.slice(start=-1).isin(['A', 'B', 'C'])]
```

```python
stage_sliced = stage_to_slice.astype(str).str.slice(start=0, stop=-1)
stage = pd.concat([stage_no_need_to_slice , stage_sliced])
```

```python
stage.value_counts().sort_index() / len(stage) * 100
```
